```python
import pandas as pd
import seaborn as sns
```

```python
df_results_all = pd.read_csv('readings_for_analysis.csv', sep='\t')

first = df_results_all.first_reading.dropna().str.split(pat=';', expand=True)[2]
first = first.str.split(pat=':', expand=True)[0].astype(int) * 60 + first.str.split(pat=':', expand=True)[1].astype(int)
second = df_results_all.second_reading.dropna().str.split(pat=';', expand=True)[2]
second = second.str.split(pat=':', expand=True)[0].astype(int) * 60 + second.str.split(pat=':', expand=True)[1].astype(int)
third = df_results_all.third_reading.dropna().str.split(pat=';', expand=True)[2]
third = third.str.split(pat=':', expand=True)[0].astype(int) * 60 + third.str.split(pat=':', expand=True)[1].astype(int)
```

```python
sns.lineplot(data=first).set(title='First reading', xlabel='Číslo batchu', ylabel='Čas v sekundách')
print("Priemer: ", round(first.mean(), 2), "\nPocet batchov: ", first.count())
```

```python
sns.lineplot(data=second).set(title='Second reading', xlabel='Číslo batchu', ylabel='Čas v sekundách')
print("Priemer: ", round(second.mean(), 2), "\nPocet batchov: ", second.count())
```

```python
sns.lineplot(data=third).set(title='Third reading', xlabel='Číslo batchu', ylabel='Čas v sekundách')
print("Priemer: ", round(third.mean(), 2), "\nPocet batchov: ", third.count())
```
